```python
import pandas as pd
import numpy as np

import json
```

```python
def get_top5_tables(store_data):
    df = pd.DataFrame.from_dict(store_data, orient='index')
    df = df[['Net Revenue', 'Total Cost of Sales', 'EBITDA']].fillna(0)
    df = df[df['Net Revenue'] > 0]

    # Sort and take top 5
    top_revenue = df['Net Revenue'].sort_values( ascending=False).head(5).reset_index().rename(columns={'index': 'Store', 'Net Revenue': 'Top 5 Net Revenue'})
    top_cogs = df['Total Cost of Sales'].sort_values( ascending=False).head(5).reset_index().rename(columns={'index': 'Store', 'Total Cost of Sales': 'Top 5 Total Cost of Sales'})
    top_ebitda = df['EBITDA'].sort_values( ascending=False).head(5).reset_index().rename(columns={'index': 'Store', 'EBITDA': 'Top 5 EBITDA'})


    return top_revenue, top_cogs, top_ebitda
```

```python
with open('Data/all_store_data_enriched.json', 'r') as f:
    all_store_data = json.load(f)
```

```python
store_data = extract_all_line_items(all_store_data,'P10-24','All')
```
